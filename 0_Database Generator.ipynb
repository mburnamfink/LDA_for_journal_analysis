```python
import pymongo as pm
import metaknowledge as mk
import os
import pyprind
```

This notebook initializes the MongoDB database from Web of Science full record + cites collection objects, which are stored in a scrapes/{journal} directory structure.

Metaknowledge is a python package for bibliographic analysis created by Reid McIlroy-Young, John McLevey, and Jillian Anderson, and is available at http://networkslab.org/metaknowledge/

```python
client = pm.MongoClient()
print(client.HOST)
print(client.PORT)
print(client.address)
print(client.list_database_names())

local = client.local
local

local.list_collection_names()

local.drop_collection('fletcher')
local.create_collection('fletcher')

fletcher = local.get_collection('fletcher')

#os.chdir('..')

print(os.getcwd())

os.chdir('scrapes')
print(os.getcwd())

journals = os.listdir()
journals.remove('scrapes.zip')

print(len(journals))
```

```python
UTs = set()

for journal in journals:
    os.chdir(journal)
    rc= mk.RecordCollection(".")
    progbar = pyprind.ProgBar(len(rc), title=journal)
    for r in rc:
        keys = list(r.keys())
        paper = dict()
        for key in keys:
            paper[key] =  r[key]
        cites =[]
        try:
            for item in paper['CR']:
                cites.append([item.ID()+', ' + item.Extra()])
        except:
            pass
        paper['CR'] = cites
        try:
            UT = paper['UT']
            if UT not in UTs:
                fletcher.insert_one(paper)
                UTs.add(UT)
        except:
            pass

        progbar.update()
    os.chdir('..')
```
